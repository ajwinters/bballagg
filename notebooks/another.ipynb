```python
# Import required libraries
from nba_api.stats.endpoints import commonplayerinfo
import pandas as pd
pd.set_option("display.max_columns", None)  # Show all columns
# Pull CommonPlayerInfo endpoint into a DataFrame
# Using LeBron James as an example (Player ID: 2544)
player_info = commonplayerinfo.CommonPlayerInfo(
    player_id='2544'  # LeBron James
)

# Get the DataFrame
player_info_df = player_info.get_data_frames()[0]

print(f"Player info retrieved for player ID: 2544")
print(f"Columns: {list(player_info_df.columns)}")
player_info_df
```

```python
# Import and use CommonTeamRoster endpoint
from nba_api.stats.endpoints import commonteamroster

# Get team roster for Lakers (Team ID: 1610612747)
team_roster = commonteamroster.CommonTeamRoster(
    team_id='1610612747',  # Lakers
    season='2024-25'
)

# Get the roster DataFrame
roster_df = team_roster.get_data_frames()[0]

print(f"Team roster retrieved for Lakers (Team ID: 1610612747)")
print(f"Number of players: {len(roster_df)}")
print(f"Columns: {list(roster_df.columns)}")
roster_df
```

```python
# Import and use CommonAllPlayers endpoint
from nba_api.stats.endpoints import commonallplayers

# Get all players for the 2024-25 season
all_players = commonallplayers.CommonAllPlayers(
    league_id='00',  # NBA
    season='2024-25',
    is_only_current_season='0'  # Get all players who played in that season
)

# Get the players DataFrame
all_players_df = all_players.get_data_frames()[0]

print(f"All players retrieved for 2024-25 season")
print(f"Total players found: {len(all_players_df)}")
print(f"Columns: {list(all_players_df.columns)}")
all_players_df
```
